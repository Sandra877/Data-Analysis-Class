```python
#the imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#reading the content of our csv file
path = "/content/drive/MyDrive/data sets/Census.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
#to do a descriptive analysis of the data
df.describe()
```

```python
#to calculate the limits of the data in order oto get the outliers
upper_limit = df['Family Composition'].mean() + 3*df['Family Composition'].std()
lower_limit = df['Family Composition'].mean() - 3*df['Family Composition'].std()
print(upper_limit)
print(lower_limit)
```

```python
#to get the outliers
df.loc[(df['Family Composition'] < upper_limit) | (df['Family Composition']> lower_limit)]
```

```python
#histogram of family composition
plt.hist(df['Family Composition'], bins=45, color='blue', edgecolor='black')
plt.title('Histogram of Family Composition')
plt.xlabel('Frequency')
plt.ylabel('family composition')
plt.show()
```

```python
#scatter plot to show whether the gender has as aneffect on economic activity
plt.scatter(df['Sex'], df['Economic Activity'], color='red', alpha=0.5)
plt.title('Scatter Plot sex, Economic Activity')
plt.xlabel('sex')
plt.ylabel('Economic Activity')
plt.show()
```

```python
#bar chart to visualize the number of hours worked per week by the population
df['Hours worked per week'].value_counts().plot(kind='bar', color='green')
plt.title('Bar Chart for Hours')
plt.xlabel('No of hours')
plt.ylabel('frequency of each set of hours')
plt.show()
```

```python
 #heat map to show the relationship btn region and family composition
 #Create a pivot table to count occurrences of each family composition within each Region
pivot_table = pd.pivot_table(df, index='Family Composition', columns='Region', aggfunc='size', fill_value=0)

# Plot a heatmap
plt.figure(figsize=(10, 4))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g', cbar_kws={'label': 'Count'})
plt.title('Relationship between Region and Family Composition')
plt.xlabel('Region')
plt.ylabel('Family Composition')

plt.show()
```

```python
#pie chart to show the ratio between the student and non student in the population
student_counts = df['Student'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(student_counts, labels=student_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Student ')

# Show the pie chart
plt.show()
```

```python
#box plot to show distribution of Age across Different Regions
plt.figure(figsize=(15, 5))
plt.title('Distribution of Age across Different Regions')
plt.xlabel('Region')
plt.ylabel('Age')

# Using seaborn for a boxplot
sns.boxplot(x='Region', y='Age', data=df)

plt.show()
```

```python
#bar chart to show the distribution of the marital status among the population
# Count the occurrences of each Marital Status
marital_status_counts = df['Marital Status'].value_counts()

# Plot a bar chart
plt.figure(figsize=(8, 5))
plt.bar(marital_status_counts.index, marital_status_counts.values, color='skyblue')
plt.title('Distribution of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')

plt.show()
```

```python
#heat map to show relationship between Economic Activity and Occupation
# Create a pivot table to count occurrences of each Occupation within each Economic Activity
pivot_table = pd.pivot_table(df, index='Economic Activity', columns='Occupation', aggfunc='size', fill_value=0)

# Plot a heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g', cbar_kws={'label': 'Count'})
plt.title('Relationship between Economic Activity and Occupation')
plt.xlabel('Occupation')
plt.ylabel('Economic Activity')

plt.show()
```
